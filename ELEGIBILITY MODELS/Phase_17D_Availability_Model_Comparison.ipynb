
Create comparison table

```python
import pandas as pd

comparison_df = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],
    "Accuracy": [
        0.49,
        0.4965,
        0.4725
    ],
    "Precision": [
        0.50,   # class "Yes"
        0.50,
        0.48
    ],
    "Recall": [
        0.65,
        0.81,
        0.52
    ],
    "F1_Score": [
        0.57,
        0.62,
        0.50
    ],
    "ROC_AUC": [
        0.4977,
        0.4864,
        0.4741
    ]
})

comparison_df
```

Sort by best model

```python
comparison_df.sort_values(
    by=["F1_Score", "ROC_AUC", "Accuracy"],
    ascending=False
)
```

Visual comparison

```python
import matplotlib.pyplot as plt

comparison_df.set_index("Model")[["Accuracy", "Precision", "Recall", "F1_Score", "ROC_AUC"]].plot(
    kind="bar",
    figsize=(10, 6)
)

plt.title("Availability Model Performance Comparison")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.xticks(rotation=30)
plt.show()
```

Select best model

```python
best_model = comparison_df.sort_values(
    by=["F1_Score", "ROC_AUC", "Accuracy"],
    ascending=False
).iloc[0]

best_model
```

Decision statement

```python
print(
    f"The selected availability prediction model is {best_model['Model']} "
    f"with F1-score of {best_model['F1_Score']} and ROC-AUC of {best_model['ROC_AUC']}."
)
```
