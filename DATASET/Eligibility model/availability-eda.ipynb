Load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df2 = pd.read_csv("availability_dataset.csv")

df2.head()
```

Data Shape

```python
df2.shape
```

Inspect structure

```python
df2.info()
```

Check missing values

```python
df2.isnull().sum()
```

Inspect target variable

```python
df2["availability"].value_counts()

df2["availability"].value_counts().plot(kind="bar")
plt.title("Availability Distribution")
plt.ylabel("Count")
plt.show()
```

Distribution of donation history

```python
# Months since first donation
df2["months_since_first_donation"].describe()

df2["months_since_first_donation"].hist(bins=30)
plt.title("Months Since First Donation")
plt.show()

# Number of donations
df2["number_of_donation"].describe()

df2["number_of_donation"].hist(bins=30)
plt.title("Number of Donations")
plt.show()

# Pints donated
df2["pints_donated"].describe()

df2["pints_donated"].hist(bins=30)
plt.title("Pints Donated Distribution")
plt.show()
```

Relationship between predictors and availability

```python
# Donations vs availability
pd.crosstab(df2["number_of_donation"], df2["availability"])

# Months since first donation vs availability
pd.crosstab(df2["months_since_first_donation"], df2["availability"])

# Blood group vs availability
pd.crosstab(df2["blood_group"], df2["availability"])
```

Correlation analysis

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.heatmap(df2.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
