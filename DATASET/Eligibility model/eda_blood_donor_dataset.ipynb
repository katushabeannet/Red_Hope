Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

Load the Cleaned Dataset and show Data structure

```python
df = pd.read_csv("cleaned_blood_donor_model_dataset.csv")

print("Dataset Shape:", df.shape)
df.head()

df.info()
```

Statistical Summary

```python
df.describe()
```

Check Missing Values

```python
df.isnull().sum()
```

Class Distribution (Target Variable)

```python
df["Eligible_for_Donation_binary"].value_counts()

sns.countplot(x="Eligible_for_Donation_binary", data=df)
plt.title("Eligibility Distribution")
plt.show()
```

Feature Distributions

```python
# - Hemoglobin Distribution

sns.histplot(df["Hemoglobin_g_dL"], kde=True)
plt.title("Hemoglobin Distribution")
plt.show()

# - Weight Distribution

sns.histplot(df["Weight_kg"], kde=True)
plt.title("Weight Distribution")
plt.show()

# - Age Distribution

sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

Feature Relationship with Eligibility

```python
# - Hemoglobin vs Eligibility

sns.boxplot(x="Eligible_for_Donation_binary", y="Hemoglobin_g_dL", data=df)
plt.title("Hemoglobin vs Eligibility")
plt.show()

# - Weight vs Eligibility

sns.boxplot(x="Eligible_for_Donation_binary", y="Weight_kg", data=df)
plt.title("Weight vs Eligibility")
plt.show()

# - Age vs Eligibility

sns.boxplot(x="Eligible_for_Donation_binary", y="Age", data=df)
plt.title("Age vs Eligibility")
plt.show()

# - days_since_last_donation vs Eligibility

sns.boxplot(x="Eligible_for_Donation_binary", y="days_since_last_donation", data=df)
plt.title("Donation Interval vs Eligibility")
plt.show()

```

Correlation Analysis

```python
corr = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

Binary Feature Analysis

```python
# - Medical Condition

sns.countplot(x="has_medical_condition", hue="Eligible_for_Donation_binary", data=df)
plt.title("Medical Condition vs Eligibility")
plt.show()

# - Low Hemoglobin

sns.countplot(x="low_hemoglobin", hue="Eligible_for_Donation_binary", data=df)
plt.title("Low Hemoglobin vs Eligibility")
plt.show()

# - Low Weight

sns.countplot(x="low_weight", hue="Eligible_for_Donation_binary", data=df)
plt.title("Low Weight vs Eligibility")
plt.show()
```

Pairplot (Feature Relationships)

```python
sns.pairplot(df, hue="Eligible_for_Donation_binary")
plt.show()
```
